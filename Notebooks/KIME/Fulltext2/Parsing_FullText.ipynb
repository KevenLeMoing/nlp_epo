```python
# Import du fichier de test

data = read_csv('C:\Users\k_lemo\Documents\Stage\Data\KIME\XML\importXml2\Test\docid_fulltext2.csv', header=None)
len(data)
```

```python
# Parsing des "full-text"

start_time = time.time()
n=0
data_text = DataFrame(np.zeros(0,dtype=[('sectionId', 'i4'),('Type', 'a50'),('Text', 'a50')]))
for i in range(len(data)):
    try:
        if len(data[0][i])==36:
            sectionID = data[0][i]
            root = ET.fromstring(data[1][i])
            Type = root.get('type')
            text = root[0].text 
            data_text = data_text.append({'sectionId':sectionID,'Type':Type,'Text':text},ignore_index=True)
    except:
        n+=1
data_fulltext = data_text.dropna(how='any',subset=['sectionId','Type','Text'])
data_fulltext = data_fulltext.reset_index(drop=True)
# On retire les balises html
clean = [bs.BeautifulSoup(data_fulltext['Text'][i], "lxml").get_text() for i in range(len(data_fulltext))]
data_fulltext['Text'] = clean
#On supprime tout ce qui relève de la ponctuation et du numérique
clean = [re.sub("[^a-zA-Z\xc4\xe4\xc9\xe9\xd6\xf6\xdc\xfc\xdf\xc0\xe0\xc2\xe2\xc6\xe6\xc7\xe7\xc8\xe8\xca\xea\xcb\xeb\xce\xee\xcf\xef\x8c\x9c\xd9\xf9\xdb\xfb]"," ",data_fulltext['Text'][i]) for i in range(len(data_fulltext))]
data_fulltext['Text'] = clean
p = (len(data))-(len(data_fulltext))
print(n)
print("")
print(p)
print("")
print("--- %s seconds ---" % (time.time() - start_time))
```
